```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('data.txt', skiprows=2, dtype='f8')
years = data[:10,0] #первые 10 зaпuсей
n = len(years)
fig = plt.figure()
ax = fig.add_subplot(111)
sources = ('Процессоp', 'Память', 'Плата', 'Жесткие диски')
hatch = ['00', '', 'xxxx', '//']
bottom = np.zeros(n)
bars = [None]*n
for i, source in enumerate(sources):
 bars[i] = ax.bar(years, bottom=bottom, height=data[:10,i+1], color='g',
 hatch=hatch[i], align='center', edgecolor='k')
 bottom += data[:10,i+1]
ax.set_xticks(years)
plt.xticks(rotation=90)
ax.set_xlim(2008, 2019)
ax.set_ylabel('Количество продаж')
ax.set_title('Продажи компании, 1990-2018')
plt.legend(bars, sources, loc='best')
plt.show()
plt.draw()
fig.savefig('Количество продаж.png', dpi=500)
```
